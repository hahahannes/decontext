```python
import statistics
import math
```

```python
def get_cv_score_dataframes_per_dim(results, sort_results=False):
    means = np.asarray(results).mean(axis=0)
    stds = np.asarray(results).std(axis=0)
    n_samples = len(means)

    rows = []
    for dim_id in range(49):
        mean = means[dim_id]
        stdev = stds[dim_id]
        confidence_interval = 2.575 * stdev / math.sqrt(n_samples)
        rows.append({'Dimension ID': dim_id+1, 'mean': mean, 'conf': confidence_interval})
    df = pd.DataFrame(rows)
    if sort_results:
        df = df.sort_values(by='mean', ascending=False)
    return df 
```

```python
def plot(title, ax, df, ylabel):
    values = df['mean']
    ax.bar(range(df.shape[0]), values, yerr=df['conf'])
    ax.set_title(title)
    ax.set_xticks(range(49))
    ax.set_xticklabels(df['Dimension ID'])
    ax.set_xlabel('Dimension ID')
    ax.set_ylabel(ylabel)
```

```python
def get_subplots(results):
    return plt.subplots(len(results),1, figsize=(11,35))
```

```python
fig_r2, axes_r2 = get_subplots(results)
#fig.suptitle('R2 per dimensions')

fig_corr_sim, axes_sim_correlations= get_subplots(results)
#fig.suptitle('Spearman correlation between similarity matrices based on predicted dimensions')

fig_corr_dim, axes_correlations = get_subplots(results)
#fig.suptitle('Correlation between predicted and true dimension value')

for i, _ in enumerate(results.items()):
    model, model_results = _
    
    
    r2_scores = model_results['r2_cv_scores']
    r2_scores = get_cv_score_dataframes_per_dim(r2_scores)
    ylabel = 'Coefficient of Determination'
    title = model_prettify(model)
    plot(title, axes_r2[i], r2_scores, ylabel)

    sim_correlations_per_dim = model_results['sim_correlations_per_dim']
    sim_correlations_per_dim = get_cv_score_dataframes_per_dim(sim_correlations_per_dim)
    ylabel = 'Correlation of Similarity Vectors based on Dimension (C3)'
    title = model_prettify(model)
    plot(title, axes_sim_correlations[i], sim_correlations_per_dim, ylabel)

    correlations_per_dim = model_results['correlations_per_dim']
    correlations_per_dim = get_cv_score_dataframes_per_dim(correlations_per_dim)
    ylabel = 'Correlation between predicted and true dimension values (C1)'
    title = model_prettify(model)
    plot(title, axes_correlations[i], correlations_per_dim, ylabel) 

plt.tight_layout()
plt.show()
fig_r2.savefig(pjoin(FIGURE_OUTPUT_DIR, 'dim_pred_r2.pdf'), bbox_inches='tight')
fig_corr_sim.savefig(pjoin(FIGURE_OUTPUT_DIR, 'dim_pred_correaltion_similarity_matrix.pdf'), bbox_inches='tight')
fig_corr_dim.savefig(pjoin(FIGURE_OUTPUT_DIR, 'dim_pred_correlation_dimension_values.pdf'), bbox_inches='tight')
```

# Example to visualize why two words are similar according to BERT embeddings

```python
def plot_two(bert1, bert2, human1, human2, label1, label2):
    fig, ax = plt.subplots(2,1, figsize=(15,10), sharey=True)

    n_dims = 49
    ax[0].bar(range(n_dims), bert1, label=label1, width=0.8, align='center')
    ax[0].bar(range(n_dims), bert2, label=label2, width=0.4, align='center')

    ax[1].bar(range(n_dims), human1, label=label1, width=0.8, align='center')
    ax[1].bar(range(n_dims), human2, label=label2, width=0.4, align='center')
    
    for s_ax in ax:
        s_ax.set_xticks(range(49))
        s_ax.set_xticklabels(range(1,50))
    
    ax[0].set_xlabel('Dimension ID')
    ax[0].set_ylabel('Predicted dimensions value')
    ax[0].set_title('BERT')
    ax[1].set_title('THINGS')
    ax[0].legend()
```
