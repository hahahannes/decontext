```python
def run(combs, matching):
    corpus = 'wikidumps'
    results = defaultdict(lambda: defaultdict(list))
    
    matching_words_things = get_intersection_words(1, matching=matching, corpus_folder=corpus, folder='thinga')
    print(f'use {len(matching_words_things)} for things')
    
    for model, embedding_path, static_matching in yield_static_data('thinga'):
        pearson, spearman, matrix, vector = run_evaluation(embedding_path, matching=static_matching, matching_words=matching_words_things)
        results[model] = spearman.correlation

    for model, n_context_mapping in combs.items():
        for n_context, layers in n_context_mapping.items():
            for layer in layers:
                    # TODO homonyms and not vocab words
                    #matching_words_things = remove_not_in_vocab(matching_words_things, True)


                    # TODO when using mathcing words from wikidump -> layer 12 performs way better on wikitext-103
                    # mayeb there are some words, where more context lead to better layer 12 results


                path = f'{EMBEDDING_DATA_DIR}/thinga/{corpus}/decontext/{model}/{layer}/{matching}/mean/{n_context}/decontext.txt'

                pearson, spearman, matrix, vector = run_evaluation(path, matching=matching, matching_words=matching_words_things)
                results[model][n_context].append(spearman.correlation)
    
    return results
```

```python
def plot(combs, results):
    fig, axes = plt.subplots(3,2, figsize=(10,10), sharey= True)

    for i, _ in enumerate(combs.items()):
        ax = get_ax(axes, i)
        model, n_context_mapping = _
        for n_context, layers in n_context_mapping.items():
            layers = [str(layer) for layer in layers]            
            label = 'Decontextualized' if n_context == 'all' else 'Contextualized'
            ax.plot(layers, results[model][n_context], marker='o', label=label)

        ax.set_xticks(layers)
        ax.set_title(f'{model_prettify(model)}')
        
        ax.axhline(y=results['w2v'], color='r', linestyle='--', label='Word2Vec')
        ax.axhline(y=results['glove'], color='g', linestyle='--', label='GloVe')
        ax.axhline(y=results['deconf'], color='black', linestyle='--', label='Deconflated')

    ax = axes[0][0]
    ax.legend()
    ax.set_xlabel('Extraction layer')
    ax.set_ylabel('Spearman correlation')
    ax.set_ylim([0, 0.5])

    plt.tight_layout()
    plt.savefig(f'things.pdf', bbox_inches='tight')
```

```python
from collections import defaultdict
result_filter = defaultdict(lambda: defaultdict(list))

combs = {
            'bert-base': range(13),
            'bert-large': range(25),
            'gpt-2': range(13),
            'gpt-2-medium': range(25),
            'sbert_bert': range(13),
            'sbert_distill_roberta': range(7)
}

n_unique_contexts = [1, 10, 50, 100, 300, 500]
n_words = []

coprus_folder = 'wikidumps'
matching = 'word'
matching_words = get_intersection_words(500, coprus_folder, matching, folder='thinga')

for model, layers in combs.items():
    for n in n_unique_contexts:
        for layer in layers:
            n_str = str(n)
            filter_path = f'{EMBEDDING_DATA_DIR}/thinga/{coprus_folder}/decontext/{model}/{layer}/{matching}/mean/{n_str}/decontext.txt'
            pearson, spearman, sim_matrix, sim_vec = run_evaluation(filter_path, matching, matching_words=matching_words)
            result_filter[model][layer].append(spearman.correlation)
```

```python
import numpy as np 


x_ticks = range(len(n_unique_contexts))

fig, axes = plt.subplots(3,2, sharey=True, figsize=(12,10))
for i, _ in enumerate(combs.items()):
    ax = get_ax(axes, i)
    model, layers = _
    colors = plt.cm.Spectral(np.linspace(0,1,len(layers)))
    for layer in layers:
        label = '_nolegend_'
        if len(layers) > 20:
            if layer % 5 == 0:
                label = f'Layer {layer}'
        elif len(layers) > 10:
            if layer % 3 == 0:
                label = f'Layer {layer}'
        else:
            label = f'Layer {layer}'
                    
        ax.plot(x_ticks, result_filter[model][layer], marker='o', label=label, color=colors[layer])
        ax.legend()
        ax.xaxis.set_ticks(x_ticks)
        ax.xaxis.set_ticklabels(n_unique_contexts)
        ax.set_ylabel(f'Spearman correlation')
        ax.set_title(model_prettify(model))

axes[0][0].set_xlabel('Number of contexts for aggregation')
plt.tight_layout()
plt.savefig(f'things_layer_context.pdf', bbox_inches='tight')

```

```python
from things_evaluation.evaluate import load_behav, load_sorting, match_behv_sim
import pandas as pd
behav_sim = load_behav()
sorting_df = load_sorting()
things_df = pd.DataFrame(behav_sim, index=sorting_df['concept_id'], columns=sorting_df['concept_id'])
things_df = things_df.loc[matching_words, matching_words]
things_df = things_df.reindex(sorted(things_df.columns), axis=1)
things_df = things_df.reindex(sorted(things_df.columns), axis=0)

things_df
```

```python
from collections import defaultdict
results = defaultdict(list)

def correlation_number_context(layer):
    embedding_path = f'../../data/thinga/wikidumps/decontext/bert-base/{layer}/{matching}/mean/all/decontextualized_sense_embeddings.txt'

    # similarity matrix
    p, s, n, sim_matrix = run_evaluation(embedding_path, matching, matching_words=matching_words)
    sim_matrix = sim_matrix.reindex(sorted(sim_matrix.columns), axis=1)
    sim_matrix = sim_matrix.reindex(sorted(sim_matrix.columns), axis=0)

    # number of contexts
    context_df = read_embeddings(embedding_path, matching, matching_words, None, True)


    rows = []
    for column in things_df.columns:
        all_true_sims_per_col = list(things_df[column])
        all_pred_sims_per_col = list(sim_matrix[column])
        correlation = spearmanr(all_true_sims_per_col, all_pred_sims_per_col)
        n_contexts = context_df.loc[column, 'n_contexts']
        rows.append({'concept': column, 'spearman': correlation[0], 'n_context': n_contexts})
    
    corr_df = pd.DataFrame(rows).sort_values('spearman')
    lin_reg = linregress(corr_df['n_context'], corr_df['spearman'])
    results[layer] = (corr_df, lin_reg)
```

```python
from scipy.stats import linregress

for layer in ['0', '1', '12']:
    correlation_number_context(layer)
```

```python
fig, axes = plt.subplots(1,3, figsize=(20,5), sharey=True)
import numpy as np 

for i, layer in enumerate(results):
    corr_df, lin_reg = results[layer]
    slope, intercept, r_value, p_value, std_err = lin_reg
    line_x = np.arange(corr_df['n_context'].min(), corr_df['n_context'].max())
    line_y = slope*line_x + intercept
    axes[i].plot(line_x, line_y, label='$%.2fx + %.2f$, $R^2=%.2f$' % (slope, intercept, r_value**2), c='red')
    axes[i].scatter(corr_df['n_context'], corr_df['spearman'])
    axes[0].set_ylabel('Spearman correlation')
    axes[i].set_xlabel('number of contexts')
    axes[i].set_title(f'Layer {layer}')
    fig.suptitle('Spearman correlation between the THINGS concept similarity vector and the embedding similarity vector for each concept (n=1854)')
```
