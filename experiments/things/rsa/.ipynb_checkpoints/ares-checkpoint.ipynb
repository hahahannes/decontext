```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys 
sys.path.append('../../..')
import os 
EMBEDDING_DATA_DIR = '/home/hhansen/decon/decon_env/data'
os.environ['EMBEDDING_DATA_DIR'] = EMBEDDING_DATA_DIR
os.environ['EMBEDDING_EVALUATION_DATA_PATH'] = '/home/hhansen/decon/decon_env/DecontextEmbeddings/helpers/embedding_evaluation/data/'
DATA_DIR = '/home/hhansen/decon/decon_env/DecontextEmbeddings/data'
os.environ['DATA_DIR'] = DATA_DIR


from helpers.embedding_evaluation.evaluate import Evaluation as wordsim_evaluate
from helpers.things_evaluation.evaluate import read_embeddings, load_behav, load_sorting, match_behv_sim, evaluate as run_evaluation
from helpers.data import yield_static_data, get_things_id_homonym
from helpers.intersection import get_intersection_words
from helpers.plot import get_ax, set_style_and_font_size

set_style_and_font_size()

from scipy.stats import spearmanr, pearsonr
from collections import defaultdict
import os 
```

```python
min_n_context = 10
matchings = ['word', 'synset', 'concept_id']
```

```python
word_path = f'{EMBEDDING_DATA_DIR}/thinga/wikidumps/decontext/bert-base/1/word/mean/{min_n_context}/decontext.txt'
synset_path = f'{EMBEDDING_DATA_DIR}/thinga/wikidumps/decontext/bert-base/1/synset/mean/{min_n_context}/decontext.txt'
concept_path = f'{EMBEDDING_DATA_DIR}/thinga/wikidumps/decontext/bert-base/1/concept_id/mean/{min_n_context}/decontext.txt'

matching_things_ids = None

for matching, path in [('word', word_path), ('synset', synset_path), ('concept_id', concept_path)]:
    things_df = read_embeddings(path, matching, min_n_contexts=min_n_context)
    ids = set(things_df.index)
    if not matching_things_ids:
        matching_things_ids = ids
    else:
        matching_things_ids = matching_things_ids.intersection(ids)
        
n = len(matching_things_ids)
```

```python
homonym_ids = get_things_id_homonym()
homonym_ids = set(homonym_ids).intersection(set(matching_things_ids))
n_homonym = len(homonym_ids)
homonym_ids
```

```python
results = defaultdict(list)
results_only_homonymes = defaultdict(list)


for layer in range(13):
    for matching in matchings:
        print(matching)
        embedding_path = f'{EMBEDDING_DATA_DIR}/thinga/wikidumps/decontext/bert-base/{layer}/{matching}/mean/{min_n_context}/decontext.txt'
        p, s, sim_matrix, sim_vector = run_evaluation(embedding_path, matching, matching_words=matching_things_ids)
        results[matching].append(s.correlation)
        
        p, s, sim_matrix, sim_vector = run_evaluation(embedding_path, matching, matching_words=homonym_ids)
        results_only_homonymes[matching].append(s.correlation)
```

```python
fig, axes = plt.subplots(1,2, figsize=(11,5), sharey=True)

matchings = [('word', 'Word'), ('synset', 'Word Sense'), ('concept_id', 'Synset')]

for matching, name in matchings:
    axes[0].plot(range(13), results[matching], label=f'{name} embedding')
    axes[1].plot(range(13), results_only_homonymes[matching], label=f'{name} embedding')
    
axes[0].legend()
axes[0].set_title(f'All words - n = {n}')
axes[1].set_title(f'Only homonyms - n = {n_homonym}')

plt.savefig('things_synset_word.pdf', bbox_inches='tight')
```
