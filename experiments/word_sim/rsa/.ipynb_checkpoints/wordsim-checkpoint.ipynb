```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import sys 
sys.path.append('../../..')
import os 
os.environ['DATA_DIR'] = '/home/hhansen/decon/decon_env/DecontextEmbeddings/data'

os.environ['EMBEDDING_DATA_DIR'] = '/home/hhansen/decon/decon_env/data'
os.environ['EMBEDDING_EVALUATION_DATA_PATH'] = '/home/hhansen/decon/decon_env/DecontextEmbeddings/helpers/embedding_evaluation/data/'

from helpers.embedding_evaluation.evaluate import Evaluation as wordsim_evaluate
from helpers.intersection import get_intersection_words
from helpers.data import yield_static_data
from helpers.embedding_evaluation.data import read_wordsim_embeddings
from helpers.plot import get_ax, model_prettify, set_style_and_font_size
set_style_and_font_size()

```

```python
from collections import defaultdict

simlex =  defaultdict()
wordsim =  defaultdict()

matching_words = get_intersection_words(None, 'wikidumps', 'synset', 'word_sim')

for model, embedding_path, matching in yield_static_data('word_sim'):
    embeddings = read_wordsim_embeddings(embedding_path, matching_words, None, matching, as_df=False)
    evaluation = wordsim_evaluate() 
    eval_results = evaluation.evaluate(embeddings)
    spearman_simlex = eval_results['similarity']['simlex']['all_entities']
    spearman_wordsim = eval_results['similarity']['ws353']['all_entities']
    wordsim[model] = spearman_wordsim
    simlex[model] = spearman_simlex
    
```

```python

from collections import defaultdict

print(f'use {len(matching_words)} for word sim')

combs = {
    'wikidumps': {
        'bert-base': range(13),
        'bert-large': range(25),
        'gpt-2': range(13),
        'gpt-2-medium': range(25),
        'sbert_bert': range(13),
        'sbert_distill_roberta': range(7)
    }
}

results = defaultdict(lambda: defaultdict(lambda : defaultdict(lambda: {'wordsim': [], 'simlex': []})))

for corpus_folder, mapping in combs.items():
    for model, layers in mapping.items():

        for layer in layers:    
            for n_context in ['1', 'all']:            
                path = f'../../../../data/word_sim/{corpus_folder}/decontext/{model}/{layer}/word/mean/{n_context}/decontext.txt'
                embeddings = read_wordsim_embeddings(path, matching_words, matching='word')
                evaluation = wordsim_evaluate() 
                eval_results = evaluation.evaluate(embeddings)
                spearman_simlex = eval_results['similarity']['simlex']['all_entities']
                spearman_wordsim = eval_results['similarity']['ws353']['all_entities']
                results[corpus_folder][model][n_context]['wordsim'].append(spearman_wordsim)
                results[corpus_folder][model][n_context]['simlex'].append(spearman_simlex)
       

```

```python
for dataset_name, static, key in [('Wordsim-353', wordsim, 'wordsim'), ('Simlex-999', simlex, 'simlex')]:
    fig, axes = plt.subplots(3,2, figsize=(10,10), sharey= True)

    for corpus, mapping in combs.items():
        for i, _ in enumerate(mapping.items()):
            ax = get_ax(axes, i)
            model, layers = _
            layers = [str(layer) for layer in layers]
            ax.plot(layers, results[corpus][model]['1'][key], marker='o', label='Contextualzied')
            ax.plot(layers, results[corpus][model]['all'][key], marker='o', label='Decontextualized')

            ax.axhline(y=static['w2v'], color='r', linestyle='--', label='Word2Vec')
            ax.axhline(y=static['glove'], color='g', linestyle='--', label='GloVe')
            ax.axhline(y=static['deconf'], color='black', linestyle='--', label='Deconflated')
            
            ax.set_xticks(layers)
            ax.set_title(f'{model_prettify(model)}')
            ax.tick_params(axis='y')
            ax.tick_params(axis='x')

        axes[0][0].legend()
        axes[0][0].set_xlabel('Extraction layer')
        axes[0][0].set_ylabel(f'Spearman correlation')
        axes[0][0].set_ylim([0, 0.8])

        plt.tight_layout()
        plt.savefig(f'{dataset_name}_{corpus}.pdf', bbox_inches='tight')
```

# number of context and layer

```python
from collections import defaultdict
result_wordsim = defaultdict(lambda: defaultdict(list))
result_simlex = defaultdict(lambda: defaultdict(list))

n_unique_contexts = [1, 10, 50, 100, 300, 500, 1000]

combs = {
            'bert-base': range(13),
        'bert-large': range(25),
        'gpt-2': range(13),
        'gpt-2-medium': range(25),
        'sbert_bert': range(13),
        'sbert_distill_roberta': range(7)
}

coprus_folder = 'wikidumps'
matching = 'word'
matching_words = get_intersection_words(1000, 'wikidumps', 'word', 'word_sim')

for model, layers in combs.items():
    for n in n_unique_contexts:
        for layer in layers:
            filter_path = f'../../../../data/word_sim/{coprus_folder}/decontext/{model}/{layer}/{matching}/mean/{n}/decontext.txt'
            embeddings = read_wordsim_embeddings(filter_path, matching_words, matching='word')
            evaluation = wordsim_evaluate() 
            eval_results = evaluation.evaluate(embeddings)
            spearman_simlex = eval_results['similarity']['simlex']['all_entities']
            spearman_wordsim = eval_results['similarity']['ws353']['all_entities']
            result_wordsim[model][layer].append(spearman_wordsim)
            result_simlex[model][layer].append(spearman_simlex)
```

```python
import numpy as np 


x_ticks = range(len(n_unique_contexts))

for results, name in [(result_wordsim, 'Wordsim-353'), (result_simlex, 'Simlex-999')]:
    fig, axes = plt.subplots(3,2, sharey=True, figsize=(12,10))
    for i, _ in enumerate(combs.items()):
        ax = get_ax(axes, i)
        model, layers = _
        colors = plt.cm.Spectral(np.linspace(0,1,len(layers)))
        for layer in layers:
            label = '_nolegend_'
            if len(layers) > 20:
                if layer % 5 == 0:
                    label = f'Layer {layer}'
            elif len(layers) > 10:
                if layer % 3 == 0:
                    label = f'Layer {layer}'
            else:
                label = f'Layer {layer}'
                    
            ax.plot(x_ticks, results[model][layer], marker='o', label=label, color=colors[layer])
            ax.legend()
            ax.xaxis.set_ticks(x_ticks)
            ax.xaxis.set_ticklabels(n_unique_contexts)
            ax.set_ylabel(f'Spearman correlation')
            ax.set_title(model_prettify(model))



    axes[0][0].set_xlabel('Number of contexts for aggregation')
    fig.subplots_adjust(wspace=0.2)
    plt.tight_layout()
    plt.savefig(f'{name}_layer_context.pdf', bbox_inches='tight')

#fig.suptitle(f'BERT-base - Mean aggregation - Using THINGS words - Only words with >1000 contexts - n={len(matching_words)} words total')
```
