```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial.distance import squareform
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances

import os 
os.environ['DATA_DIR'] = '/home/hhansen/decon/decon_env/DecontextEmbeddings/data'
os.environ['EMBEDDING_EVALUATION_DATA_PATH'] = '/home/hhansen/decon/decon_env/DecontextEmbeddings/helpers/embedding_evaluation/data/'

from scipy.stats import spearmanr, pearsonr
from scipy.spatial.distance import squareform
import numpy as np 
from collections import defaultdict

import os 
import sys 
sys.path.append('../..')

from helpers.things_evaluation.evaluate import evaluate, read_embeddings
from helpers.plot import get_ax
```

```python
def get_random_pairs(pairs, n):
    sample_inds = np.random.choice(np.arange(len(pairs)), size=n, replace=False)
    return np.array([pairs[i] for i in sample_inds])
```

```python
def calc_anistropy(path, n, matching):
    df = read_embeddings(path, matching=matching)
    df = df.sample(n)
    pred_sim_matrix = cosine_similarity(df, df)
    sim_vector = squareform(pred_sim_matrix, force='tovector', checks=False)
    return sim_vector.mean()
```

```python
agg = defaultdict(list)
single = defaultdict(list)
matching = 'word'
n = 500
folder = 'thinga'

combs = {
    'bert-base': range(13),
    'bert-large': range(25),
    'gpt-2': range(13),
    'gpt-2-medium': range(25),
    'sbert_bert': range(13),
    'sbert_distill_roberta': range(7),
}
for model, layers in combs.items():
    for layer in layers:
        print(f'{model} {layer}')
        decontext_path = f'../../../data/{folder}/wikidumps/decontext/{model}/{str(layer)}/{matching}/mean/all/decontext.txt'
        agg[model].append(calc_anistropy(decontext_path, n, matching))
        
        context_path = f'../../../data/{folder}/wikidumps/decontext/{model}/{str(layer)}/{matching}/mean/1/decontext.txt'
        single[model].append(calc_anistropy(context_path, n, matching))
        
        #pearson, spearman, n_words, matrix, vec= evaluate(path, matching=matching)
        #sim_vector = get_random_pairs(vec, 1000)
        #agg[model].append(np.asarray(sim_vector).mean())

        #pearson, spearman, n_words, matrix, vec= evaluate(path, matching=matching)
        #sim_vector = get_random_pairs(vec, 1000)
        #single[model].append(np.asarray(sim_vector).mean())
```

```python
fig, axes = plt.subplots(3, 2, figsize=(10,10), sharey=True)

for i, model in enumerate(agg):
    ax = get_ax(axes, i)
    ax.plot(range(len(agg[model])), agg[model], label='decontextualized')
    ax.set_title(f'{model}')
    ax.plot(range(len(single[model])), single[model], label='contextualized')

axes[0][0].legend()
axes[0][0].set_xlabel('Extraction layer')
axes[0][0].set_ylabel('Average cosine similarity')
plt.tight_layout()
plt.savefig('anistropy_aggregation.pdf', bbox_inches='tight')
```

# Anistropy Finetuning

## Sentence BERT - Finetuning on STS

```python
agg = defaultdict(list)
single = defaultdict(list)
matching = 'word'
n = 500

combs = {
    'bert-base': range(13),
    'sbert_bert': range(13)
}
for model, layers in combs.items():
    for layer in layers:
        print(f'{model} {layer}')
        decontext_path = f'../../../data/thinga/wikidumps/decontext/{model}/{str(layer)}/{matching}/mean/all/decontext.txt'
        agg[model].append(calc_anistropy(decontext_path, n, matching))
        
        context_path = f'../../../data/thinga/wikidumps/decontext/{model}/{str(layer)}/{matching}/mean/1/decontext.txt'
        single[model].append(calc_anistropy(context_path, n, matching))
```

```python
fig, axes = plt.subplots(1,1, figsize=(8,5), sharey=True)

axes.plot(range(13), agg['bert-base'], label='BERT-base: pretrained - decontextualized', c='navy')
axes.plot(range(13), agg['sbert_bert'], label='SBERT: fine tuned - decontextualized', c='red')
axes.plot(range(13), single['bert-base'], label='BERT: pretrained - contextualized', c='navy', linestyle='dotted')
axes.plot(range(13), single['sbert_bert'], label='SBERT: fine tuned - contextualized', c='red', linestyle='dotted')

plt.legend()
plt.xlabel('Extraction layer')
plt.ylabel('Average cosine similarity')
plt.savefig('anistropy_finetune.pdf', bbox_inches='tight')
```

## Own finetuning on THINGS

```python
agg_own_finetune = defaultdict(list)
single_own_finetune = defaultdict(list)

combs = {
    'bert-base': range(13)
}

file_name = 'constrastive_loss'

for model, layers in combs.items():
    for layer in layers:
        print(f'{model} {layer}')
        decontext_path = f'../../../data_fine_tune/things/wikidumps/decontext/bert-base/{str(layer)}/{matching}/mean/all/{file_name}.txt'
        agg_own_finetune[model].append(calc_anistropy(decontext_path, n, matching))
        
        context_path = f'../../../data_fine_tune/things/wikidumps/decontext/bert-base/{str(layer)}/{matching}/mean/1/{file_name}.txt'
        single_own_finetune[model].append(calc_anistropy(context_path, n, matching))
```

```python
fig, axes = plt.subplots(1,1, figsize=(8,5), sharey=True)

axes.plot(range(13), agg_own_finetune['bert-base'], label='BERT-base: finetuned - decontextualized', c='red')
axes.plot(range(13), agg['bert-base'], label='BERT-base: pretrained - decontextualized', c='navy')

axes.plot(range(13), single_own_finetune['bert-base'], label='BERT-base: finetuned - contextualized', c='red', linestyle='dotted')
axes.plot(range(13), single['bert-base'], label='BERT-base: pretrained - contextualized', c='navy', linestyle='dotted')

plt.legend()
plt.xlabel('Extraction layer')
plt.ylabel('Average cosine similarity')
plt.savefig('anistropy_own_finetune.pdf', bbox_inches='tight')
```
