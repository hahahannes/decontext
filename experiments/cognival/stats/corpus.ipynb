```python
def stats_cognival(corpus_folder):
    path = f'../../data/cognival/{corpus_folder}/decontext/bert-base/01/01/word/mean/1/decontext.txt'
    df = load_embedding_to_df(path, 'word')#[['word', 'n_contexts']]
    df = df.sort_values(by='n_contexts', ascending=False)
    df.head(30)
    total = df['n_contexts'].sum()
    print(f'{total} embeddings in total')
    print(df['n_contexts'].describe())
```

```python
def plot_hist(df_word):
    fig, axes = plt.subplots(1,1, figsize=(15,5), sharey=False, sharex=True)
    axes[0].hist(df_word['n_contexts'], bins=200)

def plot_boxplot(df_word):
    fig, axes = plt.subplots(1,1, figsize=(20,5), sharey=True, sharex=True)
    axes.boxplot([df_word['n_contexts']], vert=False)
    axes.set_yticks([1,2,3,4])
    axes.set_yticklabels(['word'])
    axes.set_xlabel('number of contexts')
```

# Wikidumps

```python
df_word, = stats_cognival('wikidumps')
```

```python
plot_hist(df_word)
plt.savefig(f'wikidumps_hist.pdf')
```

```python
plot_boxplot(df_word)
plt.savefig(f'wikidumps_boxpot.pdf')
```
