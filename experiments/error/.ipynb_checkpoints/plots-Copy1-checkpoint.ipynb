```python
def remove_not_in_vocab(matching_words):
    from transformers import BertTokenizerFast
    pretrained_weights = 'bert-base-uncased'
    tokenizer = BertTokenizerFast.from_pretrained(
                pretrained_weights, padding_side='right', model_max_length=512)

    words_not_in_vocab = []
    df = pd.read_csv('../../../../data/thinga/things_concepts.tsv', sep='\t')
    for row in df.iterrows():
        word = row[1]['Word']
        things_id = row[1]['uniqueID']
        if word not in tokenizer.vocab:
            words_not_in_vocab.append(things_id)

    return matching_words.difference(words_not_in_vocab)
```

```python
def remove_homonyms(matching_words):
    homonymns = []
    df = pd.read_csv('../../../../data/thinga/things_concepts.tsv', sep='\t')
    for row in df.iterrows():
        things_id = row[1]['uniqueID']
        if '1' in things_id or '2' in things_id or '3' in things_id:
            homonymns.append(row[1]['Word'])

    return matching_words.difference(homonymns)
```

```python
def rsa_matrix_to_pair_list(df):
    columns = list(df.columns)
    pair_rows = []

    for i, row in enumerate(df.iterrows()):
        word1 = row[0]
        for col in columns[i+1:]:
            pair_rows.append({'word1': word1, 'word2': col, 'human_pred': row[1][col]})

    return pd.DataFrame(pair_rows)
```

```python
matching = 'word'

all_matching_words = get_intersection_words(None, corpus_folder='wikidumps', folder='thinga', matching=matching)
#matching_words_tokenizer = remove_not_in_vocab(all_matching_words)
#matching_words_no_homonyms = remove_homonyms(all_matching_words)
```

```python
def get_df(things_df, matching_words):
    things_human_df = rsa_matrix_to_pair_list(things_df)

    embedding_path = f'../../../../data/thinga/wikidumps/decontext/bert-base/12/{matching}/mean/all/decontext.txt'
    p, s, n, sim_matrix, v = run_evaluation(embedding_path, matching, matching_words=matching_words)
    print('bat1' in sim_matrix.columns)
    things_decontext_df = rsa_matrix_to_pair_list(sim_matrix).rename(columns={'human_pred': 'embedding_cosine_decontext'})

    embedding_w2v_path = f'../../../../data/thinga/static/w2v/word2vec-google-news-300/embeddings.txt'
    p, s, n, sim_matrix_w2v, v = run_evaluation(embedding_w2v_path, 'word', matching_words=matching_words)
    print('bat1' in sim_matrix_w2v.columns)

    things_w2v_df = rsa_matrix_to_pair_list(sim_matrix_w2v).rename(columns={'human_pred': 'embedding_cosine_w2v'})

    things_merged_df = things_human_df.merge(things_decontext_df, on=['word1', 'word2']).merge(things_w2v_df, on=['word1', 'word2'])

    things_merged_df = get_ranks(things_merged_df)
    things_merged_df = calc_diffs(things_merged_df)

    return things_merged_df
    #import seaborn as sns 
    #sns.pairplot(things_merged_df[['human_pred', 'embedding_cosine_w2v', 'embedding_cosine_decontext']])
```

## Where do embeddings (Word2Vec, decontext embedding) predict differnt similarity than humans?

```python
def get_word_pairs_high_rank_diff(pred_df, embedding):
    if embedding == 'w2v':
        rank_diff_col = 'di^2_w2v'
        rank_col = 'embedding_rank_w2v'
        embedding_sim_col = 'embedding_cosine_w2v'
    else:
        rank_diff_col = 'di^2_decontext'
        rank_col = 'embedding_rank_decontext'
        embedding_sim_col = 'embedding_cosine_decontext'

    mismatch = pred_df.sort_values(rank_diff_col, ascending=False).head(20)
    mismatch = mismatch[['word1', 'word2', 'human_pred', embedding_sim_col]]
    print(mismatch)
```

```python
def get_word_pairs_unsimilar_to_human_similar_to_embedding(pred_df, embedding):
    if embedding == 'w2v':
        rank_diff_col = 'di^2_w2v'
        rank_col = 'embedding_rank_w2v'
        embedding_sim_col = 'embedding_cosine_w2v'
    else:
        rank_diff_col = 'di^2_decontext'
        rank_col = 'embedding_rank_decontext'
        embedding_sim_col = 'embedding_cosine_decontext'

    unsimilar_to_human_similar_to_embedding = pred_df[pred_df['human_rank'] < pred_df[rank_col]]
    mismatch = unsimilar_to_human_similar_to_embedding.sort_values(rank_diff_col, ascending=False).head(100)
    mismatch = mismatch[['word1', 'word2', 'human_pred', embedding_sim_col]]
    print(mismatch.head(20))
```

```python
def get_word_pairs_similar_to_human_unsimilar_to_embedding(pred_df, embedding):
    if embedding == 'w2v':
        rank_diff_col = 'di^2_w2v'
        rank_col = 'embedding_rank_w2v'
        embedding_sim_col = 'embedding_cosine_w2v'
    else:
        rank_diff_col = 'di^2_decontext'
        rank_col = 'embedding_rank_decontext'
        embedding_sim_col = 'embedding_cosine_decontext'

    unsimilar_to_human_similar_to_embedding = pred_df[pred_df['human_rank'] > pred_df[rank_col]]
    mismatch = unsimilar_to_human_similar_to_embedding.sort_values(rank_diff_col, ascending=False).head(100)
    mismatch = mismatch[['word1', 'word2', 'human_pred', embedding_sim_col]]
    print(mismatch)
```

```python
def get_word_pairs_similar_to_human_similar_to_embedding(pred_df, embedding):
    if embedding == 'w2v':
        rank_diff_col = 'di^2_w2v'
        rank_col = 'embedding_rank_w2v'
        embedding_sim_col = 'embedding_cosine_w2v'
    else:
        rank_diff_col = 'di^2_decontext'
        rank_col = 'embedding_rank_decontext'
        embedding_sim_col = 'embedding_cosine_decontext'

    mismatch = pred_df.sort_values(rank_diff_col, ascending=True).head(20)
    b = mismatch[['word1', 'word2', 'human_pred', embedding_sim_col]]
    print(b)
    return mismatch
```

```python
def get_word_pairs_unsimilar_to_human_unsimilar_to_embedding(pred_df, embedding):
    if embedding == 'w2v':
        rank_diff_col = 'di^2_w2v'
        rank_col = 'embedding_rank_w2v'
        embedding_sim_col = 'embedding_cosine_w2v'
    else:
        rank_diff_col = 'di^2_decontext'
        rank_col = 'embedding_rank_decontext'
        embedding_sim_col = 'embedding_cosine_decontext'

    mismatch = pred_df.sort_values(rank_diff_col, ascending=False).head(20)
    b = mismatch[['word1', 'word2', 'human_pred', embedding_sim_col]]
    print(b)
    return mismatch
```

```python
def remove_not_in_vocab_wordsim(matching_words):
    from transformers import BertTokenizerFast
    pretrained_weights = 'bert-base-uncased'
    tokenizer = BertTokenizerFast.from_pretrained(
                pretrained_weights, padding_side='right', model_max_length=512)

    words_not_in_vocab = []
    df = pd.read_csv('../../data/word_sim/word_sim_vocabulary.csv', names=['word'])
    for row in df.iterrows():
        word = row[1]['word']
        if word not in tokenizer.vocab:
            words_not_in_vocab.append(word)

    return matching_words.difference(words_not_in_vocab)
```
